# A.SCALE-INVARIANT FEATURE TRANSFORM

```python
import numpy as np
import cv2 as cv
from  matplotlib import pyplot as plt
img=cv.imread("C:\\Users\\GF LAB\\img_sp_lab\\dataset\\imgs\\ship.jpg")
plt.subplot(121)
plt.imshow(img)
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
shift=cv.SIFT_create()
kp=shift.detect(gray,None)
img=cv.drawKeypoints(gray,kp,img)
plt.subplot(122)
plt.imshow(img)
plt.show()
```

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img=cv.imread("C:\\Users\\GF LAB\\img_sp_lab\\dataset\\imgs\\ship.jpg")
sift = cv.xfeatures2d.SIFT_create()
surf=cv.xfeatures2d.SURF_create(200)
kp,des=surf.detetAndCompute(img,None)
len(kp)
print(surf.getHessianThreshold())
surf.getHessianThreshold(5000)
kp,des=surf.detetAndCompute(img,None)
len(kp)
plt.imshow(img)
plt.show()
```

```python
import numpy as np
import cv2
query_img=cv2.imread("C:\\Users\\GF LAB\\img_sp_lab\\dataset\\imgs\\ship.jpg")
train_img=cv2.imread("C:\\Users\\GF LAB\\img_sp_lab\\dataset\\imgs\\ship.jpg")
query_img_bw=cv2.cvtColor(query_img,cv2.COLOR_BGR2GRAY)
train_img_bw=cv2.cvtColor(train_img,cv2.COLOR_BGR2GRAY)
orb=cv2.ORB_create()
query_Keypoints,queryDescriptors=orb.detectAndCompute(query_img_bw,None)
train_Keypoints,trainDescriptors=orb.detectAndCompute(train_img_bw,None)
matcher=cv2.BFMatcher()
matches=matcher.match(queryDescriptors,trainDescriptors)
final_img=cv2.drawMatches(query_img,query_Keypoints,train_img,train_Keypoints,matches[:20],None)
final_img=cv2.resize(final_img,(1000,650))
while(1):
    cv2.imshow("Matches",final_img)
    if cv2.waitKey(20)& 0xFF==27:
        break
cv2.destroyAllWindows()
```
