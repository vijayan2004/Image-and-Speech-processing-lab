```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
filter=np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
image=cv2.imread("C:\\Users\\GF LAB\\img_sp_lab\\dataset\\imgs\\lady.jpg")
sobel_verti_img=cv2.filter2D(image,-1,filter)
sobel_hori_img=cv2.filter2D(image,-1,np.flip(filter.T,axis=0))
plt.subplot(121)
plt.imshow(sobel_verti_img,cmap='gray')
plt.title("Vertical edge")
plt.subplot(122)
plt.imshow(sobel_hori_img,cmap='gray')
plt.title("Horizontal edge")
plt.show()
```
