# LOW PASS FILTER AND HIGH PASS FILTER

```python
import numpy as np 
import scipy.signal as sg
import matplotlib.pyplot as plt
import librosa
from scipy.signal import butter,filtfilt
d_file="C:\\Users\\GF LAB\\Desktop\\CN LAB\\img_sp_lab\\dataset\\auds\\set1.wav"
d,sr=librosa.load(d_file,8000)
t=np.linspace(0,len(d)/sr,len(d))
plt.plot(t,d,lw=1)
plt.xlabel("time(s)")
plt.ylabel("amplitude")
plt.title('orginal speech file')
plt.grid(True)
plt.show()
b,a=sg.butter(4,2000./(sr/2.),'high')
d_fil=sg.filtfilt(b,a,d)
fig,ax=plt.subplots(1,1,figsize=(6,3))
t=np.linspace(0.,len(d)/sr,len(d))
plt.plot(t,d,lw=1)
ax.plot(t,d_fil,lw=1)
plt.xlabel("time(s)")
plt.title('highpass filter')
plt.grid(True)
plt.show()
```

```python
import numpy as np 
import scipy.signal as sg
import matplotlib.pyplot as plt
import librosa
from scipy.signal import butter,filtfilt
d_file="C:\\Users\\GF LAB\\Desktop\\CN LAB\\img_sp_lab\\dataset\\auds\\set1.wav"
d,sr=librosa.load(d_file,sr=8000)
print(d)
b,a=sg.butter(4,500./(sr/2.),'low')
d_fil=sg.filtfilt(b,a,d)
fig,ax=plt.subplots(1,1,figsize=(6,3))
t=np.linspace(0.,len(d)/sr,len(d))
plt.plot(t,d,lw=1)
ax.plot(t,d_fil,lw=1)
plt.xlabel("time(s)")
plt.title('low pass filter')
plt.grid(True)
plt.show()
```
