# CONVOLUTION OF SPEECH

```python
from scipy import signal
import numpy as np
import librosa
import matplotlib.pyplot as plt
d_file="C:\\Users\\GF LAB\\Desktop\\CN LAB\\img_sp_lab\\dataset\\auds\\set1.wav"
d1_file="C:\\Users\\GF LAB\Desktop\\CN LAB\\img_sp_lab\\dataset\\auds\\set2.wav"
d,sr=librosa.load(d_file,sr=8000)
d1,sr=librosa.load(d1_file,sr=8000)
plt.subplot(2,2,1)
t=np.linspace(0,len(d)/sr,len(d))
plt.plot(t,d,lw=1)
y=signal.convolve(d,d1)
plt.grid(True)
plt.xlabel("time(s)")
plt.ylabel("amplitude")
plt.subplot(2,2,2)
t=np.linspace(0,len(d1)/sr,len(d1))
plt.plot(t,d1,lw=1)
plt.grid(True)
plt.xlabel("time(s)")
plt.ylabel("amplitude")
plt.subplot(2,2,3)
t=np.linspace(0,len(d1)/sr,len(y))
plt.plot(t,y,lw=1)
plt.grid(True)
plt.title('convolution of signals')
plt.xlabel("time(s)")
plt.ylabel("amplitude")
plt.tight_layout()
plt.grid(True)
plt.show()

```

# CORRELATION OF SPEECH SIGNALS

```python
from scipy import signal
import numpy as np
import librosa
import matplotlib.pyplot as plt
d_file="C:\\Users\\GF LAB\\Desktop\\CN LAB\\img_sp_lab\\dataset\\auds\\set1.wav"
d,sr=librosa.load(d_file,sr=8000)
plt.subplot(2,2,1)
t=np.linspace(0,len(d)/sr,len(d))
plt.plot(t,d,lw=1)
plt.xlabel("timeS in seconds")
plt.ylabel("amplitude")
r=librosa.autocorrelate(d,max_size=10000)
plt.subplot(2,1,2)
print(r.shape)
plt.title("autocorrelation plot")
plt.xlabel("lags")
plt.ylabel("coefficient")
plt.acorr(r,maxlags=20)
print("the autocorrelation plot for the data is:")
plt.grid(True)
plt.tight_layout(pad=3.0)
plt.show()
```

```python
from scipy import signal
import numpy as np
import librosa
import matplotlib.pyplot as plt
d_file="C:\\Users\\GF LAB\\Desktop\\CN LAB\\img_sp_lab\\dataset\\auds\\set1.wav"
d1_file="C:\\Users\\GF LAB\Desktop\\CN LAB\\img_sp_lab\\dataset\\auds\\set2.wav"
d,sr=librosa.load(d_file,sr=8000)
d1,sr=librosa.load(d1_file,sr=8000)
print(d)
print(d1)
correlation=np.correlate(d,d1,mode="full")
lags=np.arange(-100,100)
print(correlation)
plt.subplot(2,2,1)
plt.title('speech file 1')
t=np.linspace(0,len(d)/sr,len(d))
plt.plot(t,d,lw=1)
plt.grid(True)
plt.xlabel("time(s)")
plt.ylabel("amplitude")
plt.subplot(2,2,2)
t=np.linspace(0,len(d1)/sr,len(d1))
plt.plot(t,d1,lw=1)
plt.grid(True)
plt.xlabel("time(s)")
plt.ylabel("amplitude")
plt.subplot(2,2,3)
plt.plot(lags,correlation[0:200])
plt.title("correaltion plot")
plt.xlabel("lags")
plt.ylabel("coefficients")
plt.tight_layout()
plt.grid(True)
plt.show()

```
